```python
text = "Tokenization is the first step in text analytics. The process of breaking down a text paragraph into smaller chunks such as words or sentences is called Tokenization."
```

```python
import re
import nltk

from nltk.tokenize import word_tokenize   # 👈 FIX (missing import)
from nltk.corpus import stopwords

# Sample text
text2 = "How to remove stop words with NLTK library in Python?"

# Load stopwords
stop_words = set(stopwords.words("english"))

# Remove punctuation
text2 = re.sub('[^a-zA-Z]', ' ', text2)

# Tokenize
tokens = word_tokenize(text2.lower())

# Remove stopwords
filtered_text = []
for w in tokens:
    if w not in stop_words:
        filtered_text.append(w)

# Output
print("Tokenized Sentence:", tokens)
print("Filtered Sentence:", filtered_text)
```

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()
e_words = ["wait", "waiting", "waited", "waits"]

print("\nStemming:")
for w in e_words:
    print(w, "->", ps.stem(w))
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

text3 = "studies studying cries cry"
tokens = word_tokenize(text3)

print("\nLemmatization:")
for w in tokens:
    print(f"Lemma for {w} is {lemmatizer.lemmatize(w)}")
```
