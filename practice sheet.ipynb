```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
y2 = np.cos(x)
```

```python
plt.figure(figsize=(10, 6))
```

```python
plt.plot(x, y, label="sin(x)", color='blue', linestyle='-')
```

```python
plt.scatter(x, y2, label="cos(x)", color='red')
```

```python
categories = ['A', 'B', 'C']
values = [3, 7, 5]
plt.bar(categories, values, color='green', alpha=0.5)

```

```python
plt.title("Matplotlib Demo")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

```python
plt.grid(True)
```
