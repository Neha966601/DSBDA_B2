```python
pt = list(map(int, input("Enter value for first line (space-separatede): ").split()))
pt1 = list(map(int, input("Enter value for second line (space-separatede): ").split()))
```

```python
plt.plot(pt, c='b', ls='--', marker='*')
plt.plot(pt, c='r', ls='--', marker='o')

plt.xlabel('point')
plt.ylabel('values')
plt.show()
```

```python
plt.plot(pt, color='blue', linestyle='--', marker='*')
plt.plot(pt, color='yellow', linestyle='--', marker='o')

plt.xlabel('point')
plt.ylabel('values')
plt.title('Line Plot')
plt.show()
```
